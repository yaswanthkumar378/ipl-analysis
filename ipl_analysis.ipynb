**CLEANING THE DATA**

**checking the duplicates**

```python
print(f"Duplicate rows:{df.duplicated().sum()}")
df = df.drop_duplicates()
print("Duplicates removed!")
```

**saving a cleaned copy**

```python
df_clean=df.copy()
print("Clean dataframe ready!")
print(df_clean.shape)
```

**ANALYSE AND FINDING THE INSIGHTS**


**top 10run scorers in the ipl history**

```python
top_batsmen=(df_clean.groupby('batter')['batsman_runs'].sum().sort_values(ascending=False).head(10))
print("Top 10 run scorers:")
print(top_batsmen)
```

**top 10 wicket takers in the ipl history**

```python
top_wicket_taker=(df_clean.groupby('bowler')['is_wicket'].sum().sort_values(ascending=False).head(10))
print(top_wicket_taker)
```

**which team scores most in death overs (16–20)**

```python
death_overs=df_clean[df_clean['over'] >= 16]
team_death=(death_overs.groupby('batting_team')['batsman_runs'].sum().sort_values(ascending=False))
print("Runs scored in death overs by team:")
print(team_death)
```

**which team scores a most runs in the powerplay(0-6)**

```python
death_overs=df_clean[df_clean['over']<=6]
team_death=(death_overs.groupby('batting_team')['batsman_runs'].sum().sort_values(ascending=False))
print("Runs scored in power play by team:")
print(team_death)
```

**top 10 highest wicket takers in the power play in entire ipl history**

```python
over=df_clean[df_clean['over']<=6]
wicket_taker=(over.groupby('bowler')['is_wicket'].sum().sort_values(ascending=False).head(10))
print(wicket_taker)
```

**VISUALIZATION**

**top 10 run scorers in ipl history**

```python
plt.figure(figsize=(16,4))
top_batsmen.plot(kind='bar',color='brown')
plt.title('Top 10 IPL run scorers (all time)',fontsize=14)
for i,v in enumerate(top_batsmen):
    plt.text(i,v+90,str(v),ha='center',fontsize=9)
plt.xlabel('Total runs')
plt.savefig('top_batsmen.png', dpi=150)
plt.show()
print("Chart saved!")
```

**top 10 highest wicket takers in the ipl history**

```python
plt.figure(figsize=(25,5))
top_wicket_taker.plot(kind='bar',color='grey',edgecolor='black',fontsize=10)
for i,v in enumerate(top_wicket_taker):
    plt.text(i,v+5,str(v),ha='center',fontsize=9)
plt.title("Top 10 highest wicket takers in the ipl history",fontsize=14)
plt.xlabel("TOTAL WICKETS",fontsize=12)
plt.tight_layout()
plt.savefig('top_wicket_taker', dpi=150)
plt.show()
```

```python
over_runs = df_clean.groupby('over')['batsman_runs'].mean()
plt.figure(figsize=(12,5))
plt.plot(over_runs.index,over_runs.values,marker='o',color='green')
plt.title('Average runs per ball across all 20 overs',fontsize=14)
plt.xlabel('Over number')
plt.ylabel('Avg runs per ball')
plt.grid(True,alpha=0.3)
plt.tight_layout()
plt.savefig('runs_per_over.png', dpi=150)
plt.show()
```
